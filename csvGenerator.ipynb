```python
import csv
import shutil
import os

def copy_files_from_csv(csv_file, source_dir, dest_dir):

    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            filename1 = row[2]
            filename2 = row[3]
            filename = os.path.join(filename2, filename1)
            source_path = os.path.join(source_dir, filename)
            dest_path = os.path.join(dest_dir, filename2)
            
            if os.path.exists(source_path):
                if not os.path.exists(dest_path):
                    os.makedirs(dest_path)
                dest_path = os.path.join(dest_dir, filename)
                shutil.copy(source_path, dest_path)
            else:
                print(f"File '{filename}' not found in source directory.")

csv_file_path = "annotated.csv"
source_directory = "D:\OfficialBBBC021\BBBC021"
destination_directory = "D:\OfficialBBBC021\Annotated_subset"
copy_files_from_csv(csv_file_path, source_directory, destination_directory)
```
